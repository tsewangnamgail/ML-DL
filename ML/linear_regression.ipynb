```python
import csv

# Data to be written into the CSV file
data = [
    ['area', 'price'],   # Header row
    [2600, 550000],      # Row 1
    [3000, 565000],      # Row 2
    [3200, 610000],      # Row 3
    [3600, 680000],      # Row 4
    [4000, 725000]       # Row 5
]

# Writing to a CSV file
with open('real_estate.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("CSV file 'real_estate.csv' created successfully.")
```

```python
import pandas as pd

df = pd.read_csv('real_estate.csv')
print(df)
```

```python
from sklearn.linear_model import LinearRegression

X = df[['area']]     # Independent variable (2D)
y = df['price']      # Dependent variable

model = LinearRegression()
model.fit(X, y)

print("Model trained.")
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```

```python
area = 3500
predicted_price = model.predict([[area]])

print(f"Predicted price for {area} sq ft = ${predicted_price[0]:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(df.area, df.price)               # Actual data
plt.plot(df.area, model.predict(X))          # Regression line
plt.xlabel("Area (sq ft)")
plt.ylabel("Price ($)")
plt.title("Linear Regression - Real Estate Pricing")
plt.show()
```
