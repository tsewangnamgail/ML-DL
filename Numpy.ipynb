#axis

```python
# 1D=(axis=0)only
# 2D=(axis=0(rowa))and (axis=1(coulmn))
```

```python
x=[[1,2,3],[4,5,6],[7,1,0]]
arr=np.array(x)
arr
```

```python
print(arr.sum(axis=0))
print(arr.sum(axis=1))
```

```python
arr.T   #it will make my array trandspoce
```

```python
arr.flat
for item in arr.flat:
  print(item)
```

```python
arr.ndim  #it shows number of dimensioin
```

```python
arr.nbytes     #total bytes consume
```

```python
print(arr.argmin())
print(arr.argmax())
```

```python
arr
```

```python
print(arr.argmax(axis=0))  #max in column
print(arr.argmin(axis=1)) #max in rows
```
