#MATPLOTLIB

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
# ploting using matplotlib
# plt scatter
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('GRAPH in 2D')
plt.savefig('test.png')  # for saving the graph
```

```python
y=x*x
```

```python
plt.plot(x,y,'y--*')
```

```python
#other way
plt.plot(x,y,'r*',linestyle='dashed',linewidth=3,markersize=12)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph')
```

```python
# subplot
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'yo')
```

```python
x=np.arange(1,11)
y=3*x+4
plt.title("graph")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
```

```python
np.pi
```

```python
#sin curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sin graph')
plt.plot(x,y)
```

```python
# SUBPLOTS
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

# setting up the subplot axis
#first plot
plt.subplot(2,2,1)
plt.plot(x,y_sin,'r--')
plt.title('sin')

#second graph
plt.subplot(2,2,2)
plt.plot(x,y_cos,'g--')
plt.title('cos')
```

```python
# barplot
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('bar graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

#PICAHRT

```python
# DATA TO PLOT
labels='pythom','c++','ruby','java'
sizes=[215,130,245,210]
colors=['red','yellow','green','blue']
explode=(0.4,0,0,0)   #0.4 will move (split)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
```
