```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(np.arange(0,20).reshape(5,4),index =['r1','r2','r3','r4','r5'],columns =['c1','c2','c3','c4'])
df.head()
```

```python
df.to_csv('Test1.csv')
```

there are two ways of acessing a data

```python
#(1)  .loc
print(df.loc['r1'])
print(type(df.loc['r1']))  #series type
```

```python
#(2)  .iloc
print(df.iloc[0:2,0:2])
print(type(df.iloc[:,:]))   #data frame
```

```python
#convert data frane into arrays
df.iloc[:,:].values
```

```python
# checking null values
df.isnull().sum()
```

```python
#count unique values
print(df['c1'].value_counts())   #count number of values
print(df['c1'].unique())         #shows only unique values
```

```python
print(df['c1'])
print(type(df['c1']))  #it is of series type
print(df[['c1','c2']])
print(type(df[['c1','c2']]))# it is of data frame type
```
